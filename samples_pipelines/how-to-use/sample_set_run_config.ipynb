```python
from azureml.core import Workspace
from azureml.core.compute import AmlCompute, ComputeTarget
from azureml.pipeline.wrapper import Module, Pipeline, dsl
```

```python
workspace = Workspace._get_or_create(name="onebox-test-ws-2",
                                    subscription_id='589c7ae9-223e-45e3-a191-98433e0821a9',
                                    resource_group='onebox-test',
                                    sku='enterprise',
                                    location="centraluseuap")
print(workspace.name, workspace.resource_group, workspace.location, workspace.subscription_id, sep='\n')                   
```

```python
@dsl.pipeline(name='Hdi_pipeline', description='A pipeline with hdi module step',
              default_datastore='hdipplblobstore', default_compute_target='arthfahdi-sp')
def hdi_pipeline():
    module_func = Module.load(workspace, namespace='microsoft.com/office', name='SparkDataFetcher')
    step = module_func(
        input_extraction_path='wasbs://55d3943af884453aa7bbb6f06b8f393c'
                              '@deststoragemor6lekgfesrg.blob.core.windows.net/raw',
        category='Consumer',
        dataset='ReplyPair',
        split_on_which_date='reply',
        percent_skip_before_take='0',
        percent_take='98',
        columns='msg_Id,reply_Id,msg_Sender,reply_Sender,msg_ToRecipients,reply_ToRecipients,msg_UniqueBody,'
                'reply_UniqueBody,msg_Subject,reply_Subject,msg_CreatedDateTime,reply_CreatedDateTime',
        convert_to_adl_schema='1',
        should_persist='0',
        out_file_type='json',
        output_path='wasbs://55d3943af884453aa7bbb6f06b8f393c@deststoragemor6lekgfesrg.blob.core.windows.net/'
                    'office_migration_test/output'
    )
    conf = '{"spark.yarn.maxAppAttempts":"1",' \
           '"spark.yarn.appMasterEnv.PYSPARK_PYTHON":"/usr/bin/anaconda/envs/py35/bin/python3",' \
           '"spark.yarn.appMasterEnv.PYSPARK_DRIVER_PYTHON":"/usr/bin/anaconda/envs/py35/bin/python3"}'
    step.runsettings.configure(
        driver_memory='16g',
        driver_cores=4,
        executor_memory='3g',
        executor_cores=3,
        conf=conf)
```

```python
experiment_name = 'qianwa_test_hdi_module_new'
pipeline = hdi_pipeline()
result = pipeline.validate()
print(result)
```
